```python
import requests
from bs4 import BeautifulSoup
from random import choice
import uuid
```

```python
userAgentList = [
  'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.1.1 Safari/605.1.15',
  'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0',
  'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
  'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:77.0) Gecko/20100101 Firefox/77.0',
  'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
]
```

```python
{"User-Agent":choice(userAgentList)}
```

```python
url="https://www.ebay.com/b/Sony/bn_21835731"
```

```python
response = requests.get(url, headers={"User-Agent":choice(userAgentList)})
```

```python
response
```

```python
amazon_soup = BeautifulSoup(response.text, "html.parser")
```

```python
amazon_soup
```

```python
all_image_divs = amazon_soup.findAll('div', {"class": "b-tile__img"})
```

```python
image_srcs = [
    img.get('src')
    for i in all_image_divs
    for img in i.findChildren('img')
]
```

```python
image_srcs
```

```python
title = 0
for url in image_srcs:
    img_blob = requests.get(url, timeout=5).content
    with open('./images/' + str(uuid.uuid1()), 'wb') as img_file:
        img_file.write(img_blob)
    title+=1
```
